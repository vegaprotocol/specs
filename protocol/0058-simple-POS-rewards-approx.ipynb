```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
# minimum number of validators
minVal = 7

# competition level parameter
compLevel = 1.1

# actual number of validators
numVal = 13

def validatorScore(validatorTotalTokens):
    a = max(minVal, numVal/compLevel)
    return np.sqrt(a*validatorTotalTokens/3.0) - np.power(np.sqrt(a*validatorTotalTokens/3.0), 3.0)



xVals = np.linspace(0,0.3,100)
plt.plot(xVals,validatorScore(xVals))
plt.xlabel("input to validatorScore")
plt.ylabel("score")
```

```python
def approxOfSqrtAtOne(x):
    return 1.0 + 0.5*(x-1.0) - 0.25 * 0.5 * (x-1.0) * (x-1.0) + (3.0/8.0)*(1.0/6.0) * (x-1.0) * (x-1.0) * (x-1.0) - (15.0/16.0) * (1.0/24.0) * (x-1.0) * (x-1.0) * (x-1.0) * (x-1.0) + (105.0/32.0) * (1.0/(5*24.0)) * (x-1.0) * (x-1.0) * (x-1.0) * (x-1.0) * (x-1.0) - (945.0/64) * (1.0/(6*5*24.0)) * (x-1.0) * (x-1.0) * (x-1.0) * (x-1.0) * (x-1.0) * (x-1.0)
```

```python
xVals = np.linspace(0,2.5,100)
plt.plot(xVals,approxOfSqrtAtOne(xVals))
plt.plot(xVals,np.sqrt(xVals) )
```

```python
def validatorScoreApprox(validatorTotalTokens):
    a = max(minVal, numVal/compLevel)
    if validatorTotalTokens <= 0.4: 
        return approxOfSqrtAtOne(a*validatorTotalTokens/3.0) - np.power(approxOfSqrtAtOne(a*validatorTotalTokens/3.0), 3.0)
    else:
        return validatorScoreApprox(0.4)
```

```python
N = 100
xVals = np.linspace(0,0.9,N)
plt.plot(xVals,validatorScore(xVals))
approxVals = np.zeros(N)
for i in range(0,N):
    approxVals[i] = validatorScoreApprox(xVals[i])
plt.plot(xVals,approxVals)
```
